```python
import pandas as pd
import requests
```

```python
header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.64 Safari/537.36' }
url = 'https://br.investing.com/equities/brazil'
request = requests.get(url,headers=header)

df = pd.read_html(request.text)
```

```python
df
```

```python
for i , tabela in enumerate(df):
  print('-----------------------------')
  print(i)
  print(tabela)
```
